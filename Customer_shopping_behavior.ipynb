```python
#we will fill the missing values with median , wich is thebest option , and robust to outliers
data["Review Rating"]=data.groupby('Category')["Review Rating"].transform(lambda x:x.fillna(x.median()))

```

```python
data.isnull().sum()
```

```python
data.columns=data.columns.str.lower()
data.columns=data.columns.str.replace(" ","_")
```

```python
data.columns
```

```python
data=data.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
data.columns
```

```python
#feature engineering
#creation a new columns
label=['young_adult','adult','middle aged','senior']
data['age_group']=pd.qcut(data['age'],q=4,labels=label)
```

```python
data[['age','age_group']].head(10)
```

```python
#creacte another column
#purchase frequemcy days column
frequency_mapping={
    'Fortnightly':14,
    'Monthly':30,
    'Quarterly':90,
    'Weekly':7,
    'Annually':365,
    'Bi-Weekly':14,
    'Every 3 months':90

}
data['purchase_frequency_days']=data['frequency_of_purchases'].map(frequency_mapping)
```
